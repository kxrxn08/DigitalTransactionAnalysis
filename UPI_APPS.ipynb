```python
import requests
website="https://www.npci.org.in/what-we-do/upi/upi-ecosystem-statistics" #url
result=requests.get(website) #response
content=result.text
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(content,'html.parser')
```

```python
monthS=Month[0:3]
monthS
```

```python
tables=soup.find_all('div',id="innerTabTwo"+monthS+"23")
len(tables)
```

```python
dict1={'Application_Name':[],'Total_Volume_Mn':[],'Total_Value_Cr':[]}
```

```python
trs=tables[0].find_all('tr')
trs
```

```python
trs=trs[3:]
trs
```

```python
for i in trs:
    tds=i.find_all('td')
    dict1['Application_Name'].append(tds[1].get_text())
    dict1['Total_Volume_Mn'].append(tds[10].get_text().replace(",",""))
    dict1['Total_Value_Cr'].append(tds[11].get_text().replace(",",""))
    print(tds[1],tds[10],tds[11])
```

```python
dict1
```

```python
import pandas as pd
Apr2021=pd.DataFrame(dict1)
Apr2021
```

```python
Apr2021 = Apr2021.set_index('Application_Name')
Apr2021
```

```python
Apr2021.to_csv(Month+"_2023_UpiApps.csv")
```

```python
plt.figure(figsize=(14, 10))
for i in range(1,4):
    filename=f'Total_202{i}_UpiApps.csv'
    combined_df=pd.read_csv('./UpiAppsData/'+filename)

    grouped_df = combined_df.groupby('Application_Name').agg({
        'Total_Volume_Mn': 'sum',
        'Total_Value_Cr': 'sum'
    }).reset_index()

    # Step 3: Sort and Select the Top 10
    top_10_volume = grouped_df.sort_values(by='Total_Volume_Mn', ascending=False).head(10)
    top_10_value = grouped_df.sort_values(by='Total_Value_Cr', ascending=False).head(10)


    VolumeSortedDf = grouped_df.sort_values(by='Total_Volume_Mn',ascending=False)
    VolumeSortedDf['Percentage'] = (VolumeSortedDf['Total_Volume_Mn'] / VolumeSortedDf['Total_Volume_Mn'].sum()) * 100
    VolumeSortedDf.reset_index(drop=True,inplace=True)
    VolumeSortedDf.iloc[0:11]


    ValueSortedDf = grouped_df.sort_values(by='Total_Value_Cr',ascending=False)
    VolumeSortedDf['Percentage'] = (VolumeSortedDf['Total_Value_Cr'] / ValueSortedDf['Total_Value_Cr'].sum()) * 100
    ValueSortedDf.reset_index(drop=True,inplace=True)
    ValueSortedDf.iloc[0:11]


    top_10_volume['Percentage'] = (top_10_volume['Total_Value_Cr'] / top_10_volume['Total_Value_Cr'].sum()) * 100
    
    plt.subplot(2,3,i)
    plt.pie(top_10_volume['Percentage'][0:5],textprops={'color':"Black"},autopct='%1.2f%%',shadow=False,startangle=180)
    plt.legend(labels=top_10_volume['Application_Name'][0:5],loc="upper right")
    plt.title(f'202{i}_Top_5_Volume')

    plt.subplot(2,3,i+3)
    top_10_value['Percentage'] = (top_10_value['Total_Value_Cr'] / top_10_value['Total_Value_Cr'].sum()) * 100
    plt.pie(top_10_value['Percentage'][0:5],textprops={'color':"Black"},autopct='%1.2f%%',startangle=180)
    plt.legend(labels=top_10_value['Application_Name'][0:5],loc="upper right")
    plt.title(f'202{i}_Top_5_Value')
    plt.tight_layout()
plt.show()
```
