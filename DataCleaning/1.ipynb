```python
import requests
website="https://www.npci.org.in/what-we-do/bhim/product-statistics" #url
result=requests.get(website) #response
content=result.text
content
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
month_year=[]
No_of_Banks_live_on_BHIM=[]
Volume_in_Mn=[]
Value_in_Cr=[]
table=soup.findAll('td')
table_data=[]
for i in table:
    table_data.append(i.get_text())
table_data
length=len(table_data)
seclength=len(table_data)
while(length):
    month_year.append(table_data[seclength-length])
    length-=1
    No_of_Banks_live_on_BHIM.append(table_data[seclength-length])
    length-=1
    Volume_in_Mn.append(table_data[seclength-length])
    length-=1
    Value_in_Cr.append(table_data[seclength-length])
    length-=1
df={
    "Month":month_year,
    "No_of_Banks_live_on_BHIM":No_of_Banks_live_on_BHIM,
    "Volume_in_Mn":Volume_in_Mn,
    "Value_in_Cr":Value_in_Cr
}
import pandas as pd
df=pd.DataFrame(df)
df.to_csv('BHIM_PRODUCT_ANAYSIS_TILL_DATE.csv')
```
