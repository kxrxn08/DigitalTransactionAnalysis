```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("./Total data of digital transactions_2022.csv")
df.head()
```

```python
RTGS=df[["Unnamed: 0","RBI_Operated_RTGS_Vol","RBI_Operated_RTGS_Val"]]
RTGS
```

```python
RTGS['Rate_of_Increase_Vol'] = df['RBI_Operated_RTGS_Vol'].diff()
RTGS['Rate_of_Increase_Val'] = df['RBI_Operated_RTGS_Val'].diff()
RTGS
```

```python
RTGS.dropna(inplace=True)
RTGS.drop(index=12,inplace=True)
RTGS
```

```python
plt.figure(figsize=(10, 6))
plt.plot(RTGS['Unnamed: 0'], RTGS['Rate_of_Increase_Vol'], marker='o', linestyle='-')
plt.title('RTGS: Rate of Increaseof Volume Over Time')
plt.xlabel('Months')
plt.ylabel('Rate of Increase')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(RTGS['Unnamed: 0'], RTGS['Rate_of_Increase_Val'], marker='o', linestyle='-')
plt.title('RTGS: Rate of Increaseof Value Over Time')
plt.xlabel('Months')
plt.ylabel('Rate of Increase')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
