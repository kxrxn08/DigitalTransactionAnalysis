```python
import requests
website="https://www.npci.org.in/what-we-do/upi/upi-ecosystem-statistics" #url
result=requests.get(website) #response
content=result.text
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(content,'html.parser')
```

```python
monthS=Month[0:3]
monthS
```

```python
tables=soup.find_all('div',id="innerTabTwo"+monthS+"23")
len(tables)
```

```python
dict1={'Application_Name':[],'Total_Volume_Mn':[],'Total_Value_Cr':[]}
```

```python
trs=tables[0].find_all('tr')
trs
```

```python
trs=trs[3:]
trs
```

```python
for i in trs:
    tds=i.find_all('td')
    dict1['Application_Name'].append(tds[1].get_text())
    dict1['Total_Volume_Mn'].append(tds[10].get_text().replace(",",""))
    dict1['Total_Value_Cr'].append(tds[11].get_text().replace(",",""))
    print(tds[1],tds[10],tds[11])
```

```python
dict1
```

```python
import pandas as pd
Apr2021=pd.DataFrame(dict1)
Apr2021
```

```python
Apr2021 = Apr2021.set_index('Application_Name')
Apr2021
```

```python
Apr2021.to_csv(Month+"_2023_UpiApps.csv")
```
